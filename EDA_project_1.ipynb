```python
# outliers columns  
#1)Hours_Studied
#5)Tutoring_Sessions
#7)Exam_Score
```

```python
a=df["Hours_Studied"].describe()
q1=int(a["25%"])
q3=int(a["75%"])
iqr=q3-q1
print(a,q1,q3,iqr)
```

```python
lower_limit=q1-1.5*iqr
upper_limit=q3+1.5*iqr
```

```python
print(lower_limit,upper_limit)
```

```python
df["Hours_Studied"].clip(lower_limit,upper_limit,inplace=True)
```

```python
sns.boxplot(df["Hours_Studied"])
```

```python
a=df["Tutoring_Sessions"].describe()
q1=int(a["25%"])
q3=int(a["75%"])
iqr=q3-q1
print(a,q1,q3,iqr)
```

```python
lower_limit_1=q1-1.5*iqr
upper_limit_1=q3+1.5*iqr
```

```python
print(lower_limit_1,upper_limit_1)
```

```python
df["Tutoring_Sessions"].clip(lower_limit_1,upper_limit_1,inplace=True)
```

```python
a=df["Exam_Score"].describe()
q1=int(a["25%"])
q3=int(a["75%"])
iqr=q3-q1
print(a,q1,q3,iqr)
```

```python
lower_limit=q1-1.5*iqr
upper_limit=q3+1.5*iqr
print(upper_limit)
print(lower_limit)
```

```python
df["Exam_Score"].clip(lower_limit,upper_limit,inplace=True)
```

```python
# encoding :Encoding refers to converting categorical (non-numeric) data into a numerical format so that machine learning models can process it. 
# Categorical data can either be ordinal (with an inherent order, like "low," "medium," "high") or nominal (without an order, like "red," "green," "blue").
#  Machine learning models cannot work directly with these non-numeric categories, so encoding transforms them into numeric representations.


# feature scaling  :Feature scaling is the process of normalizing or standardizing the range of features in your data to ensure that they are on the same scale. 
# Many machine learning algorithms perform better when the input data is scaled because certain algorithms (especially distance-based ones like k-nearest neighbors or 
# gradient-based methods like neural networks) are sensitive to the magnitude of the features.
```
